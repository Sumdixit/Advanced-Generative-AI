```python
import os
print(os.environ.get("OPENAI_API_KEY"))
import openai
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,
    )
    return response.choices[0].message["content"]
print(get_completion("Hello"))

```

Level 3 – Chains (Combine Steps)

```python
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.schema import StrOutputParser

# 1. Initialize model
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)

# 2. Use ChatPromptTemplate (works with chat models)
prompt = ChatPromptTemplate.from_template(
    "Summarize the following text in one line:\n{text}"
)

# 3. Add an output parser so we get clean text
output_parser = StrOutputParser()

# 4. Chain = prompt → model → parser
chain = prompt | llm | output_parser

# 5. Run the chain
response = chain.invoke({
    "text": "LangChain helps build LLM apps by connecting prompts, models, and tools."
})
print(response)
```

```python
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.schema import StrOutputParser

# 1. Initialize the model
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)

# 2. Create a prompt template
prompt = ChatPromptTemplate.from_template(
    "Summarize the following text in one line:\n{text}"
)

# 3. Output parser
output_parser = StrOutputParser()

# 4. Chain the components: prompt → model → parser
chain = prompt | llm | output_parser

# 5. Invoke the chain with input
response = chain.invoke({
    "text": "LangChain helps build LLM apps by connecting prompts, models, and tools."
})

print(response)
```

Level 4 – Memory (Chatbot)

```python
from langchain_openai import ChatOpenAI
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
memory = ConversationBufferMemory()

conversation = ConversationChain(llm=llm, memory=memory, verbose=True)

print(conversation.invoke("Hello, I am Saurabh."))
print(conversation.invoke("What is my name?"))
```

Level 5 – Agents (Use Tools)

```python
from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent, load_tools

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
tools = load_tools(["llm-math"], llm=llm)  # gives model calculator ability

agent = initialize_agent(tools, llm, agent="zero-shot-react-description", verbose=True)

print(agent.invoke("What is 23 * 47?"))
```
